```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("sudoku.jpg",0)
img = np.float32(img)
print(img.shape)
plt.figure(), plt.imshow(img, cmap = "gray"), plt.axis("off")
```

```python
# Harris corner detection
# BlockSize: Ne kadar komusuna bakacamz belirtir.
dst = cv2.cornerHarris(img, blockSize = 2, ksize= 5, k=0.04) 
plt.figure(), plt.imshow(dst, cmap = "gray"), plt.axis("off")
```

```python
dst = cv2.dilate(dst, None) # Daha belirgin hale getirebilmek iin geniletme yntemini kullanarak kelerimizi genilettik.
img[dst>0.2*dst.max()] = 1
plt.figure(), plt.imshow(dst, cmap = "gray"), plt.axis("off")
```

```python
# shi tomsai detection
img = cv2.imread("sudoku.jpg",0)
img = np.float32(img) 
corners = cv2.goodFeaturesToTrack(img, 120, 0.01, 10) # iki ke arasndaki mesafe 10 
corners = np.int64(corners)                            

for i in corners:
    x,y = i.ravel()
    cv2.circle(img, (x,y),3,(125,125,125),cv2.FILLED)

plt.imshow(img)
plt.axis("off")
```
