```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("contour.jpg",0)
plt.figure(), plt.imshow(img, cmap = "gray"), plt.axis("off")
```

```python
# cv2.RETR_CCOMP i ve d konturlerin ikisinide bulacamz sylyor.
contours,hierarch = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE) # opencv srmne bal olarak ktlarda farkllk olabilir.
external_contour = np.zeros(img.shape) # konturleri tutacamz 2 tane matris oluturduk.
internal_contour = np.zeros(img.shape) 

for i in range(len(contours)):
    
    # External
    if hierarch[0][i][3] == -1:
        cv2.drawContours(external_contour,contours, i, 255, -1)
    else: # internal
        cv2.drawContours(internal_contour,contours, i, 255, -1)
        
plt.figure(), plt.imshow(external_contour, cmap = "gray"), plt.axis("off"),plt.title("external contour")        
plt.figure(), plt.imshow(internal_contour, cmap = "gray"), plt.axis("off"),plt.title("internal contour") 
```
