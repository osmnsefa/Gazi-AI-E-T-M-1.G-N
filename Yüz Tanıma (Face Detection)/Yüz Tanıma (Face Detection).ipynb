```python
import cv2
import matplotlib.pyplot as plt
```

```python
einstein = cv2.imread("einstein.jpg", 0)
plt.figure(), plt.imshow(einstein, cmap="gray"), plt.axis("off")
```

```python
# sınıflandırıcı 
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")  

face_rect = face_cascade.detectMultiScale(einstein)

for (x,y,w,h) in face_rect:
    cv2.rectangle(einstein, (x,y),(x+w, y+h),(255,255,255),10)
plt.figure(), plt.imshow(einstein, cmap="gray"), plt.axis("off")    
```

```python
# barcelona
barce = cv2.imread("barcelona.jpg",0)
plt.figure(), plt.imshow(barce, cmap="gray"), plt.axis("off")    
```

```python
face_rect = face_cascade.detectMultiScale(barce)

for (x,y,w,h) in face_rect:
    cv2.rectangle(barce, (x,y),(x+w, y+h),(255,255,255),10)
plt.figure(), plt.imshow(barce, cmap="gray"), plt.axis("off") 
```

```python
# barcelona
barce1 = cv2.imread("barcelona.jpg",0)
plt.figure(), plt.imshow(barce1, cmap="gray"), plt.axis("off")    
face_rect = face_cascade.detectMultiScale(barce1, minNeighbors = 7)  # minNeighbors = 7 default olarak 3'tür.Bunu değiştirerek hatayı azaltabiliriz.

for (x,y,w,h) in face_rect:
    cv2.rectangle(barce1, (x,y),(x+w, y+h),(255,255,255),10)
plt.figure(), plt.imshow(barce1, cmap="gray"), plt.axis("off") 
```

```python
# camera  (video)
cap = cv2.VideoCapture(0)

while True:
    
    ret, frame = cap.read()
    
    if ret:
        face_rect = face_cascade.detectMultiScale(frame, minNeighbors = 7)
        
        for (x,y,w,h) in face_rect:
            cv2.rectangle(frame, (x,y),(x+w, y+h),(255,255,255),10)
        cv2.imshow("face detect", frame)
        if cv2.waitKey(10) & 0xFF == ord("q"): break
cap.release() #serbest bırak
cv2.destroyAllWindows()        
        
```
